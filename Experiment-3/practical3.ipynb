```python
# List Items
# Lst items are ordered , changable , and allow duplicate values.
# List are indexed , the first item has index[0],the second has index[1] and so on'
# Allow duplicates
namelist = ["asit","dev","shiv","asit","shiv"]
print(namelist)
```

```python
# List length
namelist = ["asit","dev","shiv"]
print(len(namelist))
```

```python
# A list with strings , integers and boolean values
list1 = ["abc",69,True,False,5.6,"college"]
print(list1)
```

```python
# Access Items
list1 = ["abc",69,True,False,5.6,"college"]
print(list1[5])
print(list1[4])
```

```python
# Reverse/Negative indexing
namelist = ["asit","dev","shiv"]
print(namelist[-3])#prints index 3-3=0
print(namelist[-2])
print(namelist[-1])
```

```python
# Slicing
list2 = ["appe","banana","cherry","orange","kiwi","melon","mango","watermelon","strawberry"]
print(list2[:4])#prints from index[0] - index[4-1]
print(list2[3:8])#3 is inclusive and 8 is exclusive(i.e output till 7(8-1))
print(list2[-3:])#starts printing from [9-3=6]index till last
```

```python
# change item value
namelist=["asit","dev","shiv"]
namelist[1] = "aarush" #changed value on index[1]
namelist[0] = "kahaan" #changed value on index[0]
print(namelist)
```

```python
# Append items
namelist=["asit","dev","shiv"]
namelist.append("aarush") # added new item at the end
namelist.append("kahaan") #only one at a time
print(namelist)
```

```python
# Insert items
namelist= ["asit","kahaan","aarush"]
namelist.insert(1,"shiv")
print(namelist)
```

```python
# Extend list
namelist= ["asit","kahaan","aarush"]
namelist.extend(numberlist)
print(namelist)
```

```python
# Studend marks management system
student_marks = [10,7,8,6,9,5,4]
print(max(student_marks))
print(min(student_marks))
average_marks= sum(student_marks)/len(student_marks)
print("Average marks :", average_marks)
print(len(student_marks))
student_marks.sort()
print(student_marks)
```
