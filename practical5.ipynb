#Title -To study Sets in Python
Sets are used to store multiple items in single variable.

Set is one of 4 built-in data types in Python used to store collections of data, the other 3 are List,Tuple,and Dictionary,all with different qualities and usage. A set is a collection which is unordered,unchangeable*,and unindexed.

```python
# Creation of set
seett = {"chocolate","vanilla","pista"}
print(seett)
```

```python
#Duplicates not allowed
seett = {"chocolate","vanilla","pista","chocolate"}
print(seett)
```

```python
# The values True and 1 are considered the same value in sets
seett = {"chocolate","vanilla","pista",True,1,2}
print(seett)
```

```python

# A set with strings, integers and boolean values:
set1 = {"abc",34,True,78,"female",10.5}
print(set1)#note that in output it prints of any order as output cause set is unordered
```

```python
# To check membership
seett = {"chocolate","vanilla","pista"}

print("chocolate" in seett )
print("biscoff" in seett)
```

```python

# Once a set is created,you cannot replace its items,
# but you add new items.
seett = {"chocolate","vanilla","pista"}

seett.add("mango")
print(seett)
```

```python

#Remove Item
seett = {"chocolate","vanilla","pista"}

seett.remove("pista")

print(seett)
```

```python
# Set operations
A = {12,13,14,15,16}
B = {16,17,18,19,20}

print("Union:",A|B)
print("Intersection:",A&B)
print("Difference:",A-B)
print("Symmetric Difference:",A^B)
```

```python
# frozenset
x = frozenset({"chocolate","vanilla","pista"})
print(x)
print(type(x))
```

```python
#Adding or Removal of elements are not allowed in Frozenset
x = frozenset({"chocolate","vanilla","pista"})
x.remove("pista")
print(x)
```
